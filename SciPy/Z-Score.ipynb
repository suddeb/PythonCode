Suppose we know that the GRE scores are normally distributed with mean 157 and standard deviation 17 and IELTS scores are normally distributed with mean 6.8 and standard devialtion 1.2.

A university provides admission to their courses only on the basis of GRE and IELTS scores. The college admission committee decided to give the top performer scholarship to the student who has performed the best among all applicants. The highest score received from applicants who appeared for GRE is 182 and the highest score received from applicants who appeared for IELTS is 9.

What should be the strategy for the college to find the best candidate!

```python
# Plot the two distribution for GRE and IELTS

from scipy.stats import norm
import matplotlib.pyplot as plt
import numpy as np

fig, (ax1, ax2) = plt.subplots(1,2,figsize=(15,5))
x=np.linspace(20,300,500)
ax1.plot(x,norm.pdf(x,157,17),color="b")
ax1.set_title("GRE")
ax1.set_xlabel("GRE Score")
ax1.set_ylabel("Probability")
ax1.axvline(182,linestyle="--",color="green")

x1=np.linspace(2,10,500)
ax2.plot(x1,norm.pdf(x1,6.8,1.2),color="r")
ax2.set_title("IELTS")
ax2.set_xlabel("IELTS Score")
ax2.set_ylabel("Probability")
ax2.axvline(9,linestyle="--",color="green")
plt.show()
```

```python
# Find the Z-Score of the hihgest score in GRE and IELTS among all applicants
top_gre = (182 - 157) / 17
print("The Z-Score of the highest score in GRE among all applicants: ",top_gre);
top_ielts = (9 - 6.8) / 1.2
print("The Z-Score of the highest score in IELTS among all applicants: ",top_ielts);
```

Visualize the Z-Score

```python
fig, ax = plt.subplots()
x = np.linspace(-2,2,50)
ax.plot(x,norm.pdf(x,0,1),color="b")
ax.set_title("Standard Normal Distribution")
ax.set_xlabel("Z-Score")
ax.set_ylabel("Probability")
ax.axvline(top_gre,linestyle="--",color="green")
ax.axvline(top_ielts,linestyle="--",color="red")
plt.show()
```
