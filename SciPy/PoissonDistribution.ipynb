Create Poisson Discrete Random Variable

scipy.stats.poisson() is a poisson discrete random variable. It is inherited from the of generic methods as an instance of the rv_discrete class. 
It completes the methods with details specific for this particular distribution.

```python
# Import Necessary Libraries
from scipy.stats import poisson  
    
numargs = poisson.numargs  
a, b = 0.2, 0.8
randomVariable = poisson (a, b)  
    
print ("Random Variable : \n", randomVariable) 
```

```python
# Import Necessary Libraries
import numpy as np  
quantile = np.arange (0.01, 1, 0.1)  
  
# Random Variates  
R = poisson.rvs(a, b, size = 10)  
print ("Random Variates : \n", R)  
  
# PDF  
x = np.linspace(poisson.ppf(0.01, a, b), 
                poisson.ppf(0.99, a, b), 10) 
R = poisson.ppf(x, 1, 3) 
print ("\nProbability Distribution : \n", R) 
```

Let's Graphically Display it

```python
# Import Necessary Libraries
import numpy as np  
import matplotlib.pyplot as plt  
     
distribution = np.linspace(0, np.minimum(randomVariable.dist.b, 2))  
print("Distribution : \n", distribution)  
     
plot = plt.plot(distribution, randomVariable.ppf(distribution)) 
```
