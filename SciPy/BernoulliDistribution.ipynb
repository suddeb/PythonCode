```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def plot_bernoulli(p):
    """
    Plots the probability mass function (PMF) of a Bernoulli distribution.

    Args:
        p: The probability of success (0 <= p <= 1).
    """

    if not (0 <= p <= 1):
        raise ValueError("Probability p must be between 0 and 1.")

    # Possible outcomes (0 for failure, 1 for success)
    outcomes = [0, 1]
    probabilities = [1 - p, p]

    # Create the plot
    plt.figure(figsize=(8, 6))  # Adjust figure size for better visualization
    sns.barplot(x=outcomes, y=probabilities, palette=['skyblue', 'salmon']) 
    plt.xticks(outcomes, ['Failure (0)', 'Success (1)']) #More descriptive x-ticks
    plt.xlabel("Outcome")
    plt.ylabel("Probability")
    plt.title(f"Bernoulli Distribution (p = {p})")
    plt.ylim(0, 1) #Set y-axis limits for better visualization
    
    for i, prob in enumerate(probabilities):
        plt.text(i, prob + 0.01, f"{prob:.2f}", ha='center') #Add probability values above bars
    plt.show()

# Example usage:
p_values = [0.2, 0.5, 0.8] #Different probabilities of success
for p in p_values:
    plot_bernoulli(p)
```
