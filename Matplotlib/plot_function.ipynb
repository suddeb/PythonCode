```python
# Simple Plot
import matplotlib.pyplot as plt
import numpy as np

# X axis parameter:
xaxis = np.array([3, 10])
# Y axis parameter:
yaxis = np.array([1, 15])
plt.plot(xaxis, yaxis)
plt.show()
```

```python
# Use Marker and Linestyle
import matplotlib.pyplot as plt
import numpy as np

xaxis = np.array([3, 14, 5, 7])
# Mark each data value and customize the linestyle:
plt.plot(xaxis, marker = "X", linestyle = "--")
plt.show()
```

```python
# Scatterplot
import matplotlib.pyplot as plt
import numpy as np

# Generate random data
np.random.seed(42)
x = np.random.rand(50)
y = np.random.rand(50)

# Create scatter plot:
plt.scatter(x, y)
plt.show()
```

```python
# Multiple Datasets in One Plot
import matplotlib.pyplot as plt
import numpy as np

# Generate random data
np.random.seed(458963)

# Create random data:
xdata = np.random.random([2, 8])  

# Create two datasets from the random floats: 
xdata1 = xdata[0, :]  
xdata2 = xdata[1, :]  
# Sort the data in both datasets:
xdata1.sort()  
xdata2.sort()

# Create y data points:  
ydata1 = xdata1 ** 2
ydata2 = 1 - xdata2 ** 4

# Plot
plt.plot(xdata1, ydata1)
plt.plot(xdata2, ydata2)

plt.title("Multiple Datasets in One Plot")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
# Create a Figure with 2 rows and 2 columns of subplots:
fig, ax = plt.subplots(2, 2)
x = np.linspace(0, 5, 100)
# Index 4 Axes arrays in 4 subplots within 1 Figure: 
ax[0, 0].plot(x, np.sin(x), 'g') #row=0, column=0
ax[1, 0].plot(range(100), 'b') #row=1, column=0
ax[0, 1].plot(x, np.cos(x), 'r') #row=0, column=1
ax[1, 1].plot(x, np.tan(x), 'k') #row=1, column=1
plt.show()
```
