```python
import pandas as pd
```

```python
df = pd.read_csv('../data/itineraries.csv' , chunksize=100000)
```

```python
output = pd.DataFrame();
for chunk in df:
    categories = ['startingAirport','destinationAirport','travelDuration','segmentsAirlineName'];
    categories1 = ['startingAirport','destinationAirport'];
    details = chunk[categories]
    details['count'] = 1
    summary = details.groupby(categories).sum().reset_index();
    output = pd.concat([output,summary], ignore_index=True);
```
