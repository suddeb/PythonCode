# Interpolation

**Definition:** In mathematics, interpolation is the process of estimating unknown values between known data points. It's a technique used to **"fill in the gaps"** when you have a set of data points and want to find a value that falls within that set's range. 

**Function:** y = f(x) where for any values of x, estimate the value of y

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d
from scipy.interpolate import interp2d
```

### Example Data

```python
x_data = np.linspace(0,25,5)
y_data = x_data**3 + 5
```

### Plot the Data

```python
plt.plot(x_data, y_data,'ro:')
plt.show()
```

**Linear Interpolation** x and y are arrays of values used to approximate some function f: y = f(x). This class returns a function whose call method uses interpolation to find the value of new points.

```python
yfn = interp1d(x_data, y_data, kind='linear')
yfn(6.25)
```

Let's create many y values using this function

```python
xnew = np.linspace(0, 25, 500)
ynew = yfn(xnew)
plt.plot(x_data, y_data, 'o', xnew, ynew, '--')
plt.show()
```
